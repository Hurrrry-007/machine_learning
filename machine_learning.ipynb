## <span id="1" size=5>一、基本概念</font></span>

* 1.学习算法（learning algorithm）<br>
    * 假设用P来评估计算机程序在某类任务T上的性能，若一个程序通过利用经验E在T中任务上获得了性能改善，则关于T和P，该程序对E进行了学习。
* 2.样本维数（demensionality）:每个样本的特征数量。
* 3.监督学习（supervised learning）
    * 3.1 分类（classification）:预测结果为离散值。
        * (1)二分类（binary classification）:分为正（positive）负（negative）两类。
        * (2)多分类（multi—class classification）:分为多个类别。
    * 3.2. 回归（regression）:预测结果为连续值。
* 4.无监督学习（unsupervised learning）
    * 4.1 聚类（clustering）:将数据分为若干簇（cluster）
* 5.泛化能力（generalization）:模型适用于新样本集的能力。
* 6.独立同分布（i.i.d.）:每个样本从一个分布中独立采集得到。
* 7.版本空间（vision space）:与训练集一致的假设集合。
* 8.归纳偏好（inductive bias）:机器学习算法在学习过程中对某种类型假设的偏好。
    * 8.1 奥卡姆剃刀（Occam's razor）:若有多个假设与观察一致，选择最简单的一个。
* 9.NFL定理（No Free Lunch Theorem）:若所有问题同等重要，则所有学习算法的期望性能都相同。
    * 算法优劣只取决于具体问题。
* 10.数据决定模型上限，算法让模型无限逼近上限。

## <span id="2" size=5>二、模型评估与选择</font></span>

* 1.测试集（testing set）选择
    * 1.1 留出法(hold-out):多次随机使用分层采样法直接将数据集分为训练集（training test）和测试集两部分进行训练和测试，结果取平均。
        * 分层采样法:保留类别比例的采样方式
    * 1.2 交叉验证法（cross validation）:多次随机通过分层采样法将数据集分为K份，依次使用K-1份作为训练集，剩下的一份作为测试集，每次可以进行K次训练和测试。
      
    * 1.3 自助法（bootstrapping）:数据集D有m个样本，每次随机从数据集D中有放回的挑选一个样本放入D'，进行m次，D'将有m个样本，用D'作为训练集，D-D'作为测试集，这样的测试结果称为“包外估计”（out-of-bag-estimate）。
    * 1.4 总结:数据集小的时候用自助法，数据集足够大的时候用留出法与交叉验证法。
* 2.参数
    * (1)超参数:算法本身的参数
    * (2)模型参数。
* 3.调参
    * 验证集（validation set）:将数据集划分为训练集、验证集和测试集三部分。训练集训练模型，验证集进行模型选择和调参，测试集测试模型泛化能力。
* 4.模型评估
    * 4.1 错误率（error rate）
    * 4.2 精度（accuracy）
    * 4.3 查准率（precision）:预测真正例/预测总正例
    * 4.4 查全率（recall）:预测真正例/真实总正例；查准率与查全率通常矛盾。
      
* 5.PR曲线
    * (1)根据预测结果对样本正例进行可能性排序，按顺序逐个把样本作为正例进行预测，每次可以计算出当前P、R，P作纵轴、R作横轴。
    * (2)若A模型的PR曲线完全包裹B模型的PR曲线，则A模型优于B模型。
* 6.平衡点（Break-Even Point\BEP）:查准率=查全率的点。
    * 平衡点越大，模型越优
